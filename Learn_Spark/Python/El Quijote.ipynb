## El Quijote

Antes de nada, descargamos el archivo de  https://gist.github.com/jsdario/6d6c69398cb0c73111e49f1218960f79 y descomprimimos el zip. Una vez descomprimido, subimos el archivo a dataframe con File --> Upload Data y apuntamos la ruta a la que se sube dicho archivo. Por último, para leer el texto utilizamos la funcion **spark.read.text()**, la cual permite leer textos en Spark

```
quijote=spark.read.text("/FileStore/shared_uploads/maria.puche@bosonit.com/el_quijote-4.txt")
```

**1) Contar el número de filas que tiene nuestro texto**

```
quijote.count()
```

**2) Mostar las 20 primeras filas con los valores cortados (por defecto muestra 15 carácteres por fila, truncate=true)**

```
quijote.show()
```

**3) Mostrar las 5 primeras filas enteras (truncate = false)**

```
quijote.show(5, truncate=False)
```

**4) Diferencia entre take(), head() y first()**

Take siempre devuelve un parametro, mientras que head solo devuelve un array si le pones el parametro, sino le pones parámetro te duevuelve un row. First siempre te devuelve un row.

```
# Mostrar la primera fila, sino introducimos parámetro devuelve solo la primera fila. (devuelve un row)
quijote.head(5)
```

```
# Devolver las 5 primeras filas (siempre tiene que llevar )
quijote.take(5)
```

```
# Devolver el primer elememnto de un dataframe 
quijote.first()
```
